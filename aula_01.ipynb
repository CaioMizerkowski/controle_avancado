$$G(S)=\frac{e^{-3S}}{(S+1)^8}$$

```python
import numpy as np
import seaborn as sns
import scipy as sym
from matplotlib import pyplot as plt
import control as ctrl
from control.matlab import *
```

```python
s = TransferFunction.s;
G = ((2-3*s)/(2+3*s))/(s + 1)**8
t = 25
out, T = step(G, T=t)
plt.figure(figsize = (25,10))
p = sns.lineplot(x=T, y=out);
p.set_xticks(np.array(range(t*2))/2)
p.set_yticks(np.array(range(50))*0.02)
plt.grid(which='major')
```
